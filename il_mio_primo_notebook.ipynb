```python
# ANALIZZIAMO IL NOSTRO FILE
print("\nüìä STATISTICHE FILE")
print("=" * 25)

with open("test.txt", "r", encoding="utf-8") as file:
    contenuto = file.read()

# Analisi base
righe = contenuto.split('\n')
parole = contenuto.split()
caratteri = len(contenuto)

print(f"üìÑ Nome file: test.txt")
print(f"üìè Caratteri totali: {caratteri}")
print(f"üìù Righe totali: {len(righe)}")
print(f"üî§ Parole totali: {len(parole)}")

# Righe non vuote
righe_piene = [r for r in righe if r.strip()]
print(f"üìã Righe non vuote: {len(righe_piene)}")

# Riga pi√π lunga
riga_lunga = max(righe, key=len)
print(f"üìè Riga pi√π lunga: '{riga_lunga}' ({len(riga_lunga)} caratteri)")
```
